Step 2: Oddiy string tozalash
Har bir string columnning bo‘sh joylarini olib tashlash (strip).
Bo‘sh stringlarni NaN yoki Null bilan almashtirish.

```python
string_cols = df.select_dtypes(include='object').columns.tolist()
print(f'String columnlar ({len(string_cols)} ta):')
print(string_cols)
```

```python
for col in string_cols:
    before = df[col].dropna().str.startswith(' ').sum() + df[col].dropna().str.endswith(' ').sum()
    df[col] = df[col].str.strip()
    print(f'  {col:15s}: {before} ta bo\'sh joy olib tashlandi')

print('\n Barcha string columnlar strip qilindi')
```

Step 3: Raqam va sanalarni tozalash
Age, score, GPA, attendance, money_spent kabi numeric columnlarni tozalash va int yoki float formatiga keltirish.
Date columnlarni pandas datetime formatiga o‘tkazish.

```python
num_cols = ["age", "score", "gpa", "attendance", "money_spent"]

for col in num_cols:
    if col in df.columns:
        df[col] = pd.to_numeric(df[col], errors="coerce")

# date
if "date" in df.columns:
    df["date"] = pd.to_datetime(df["date"], errors="coerce")
```

Step 4: Email va phone validation
Emaillarni kichik harflarga o‘tkazish va noto‘g‘ri formatdagi emaillarni aniqlash.
Telefon raqamlarni tozalash va yagona standart formatga keltirish (masalan: +998...).

```python
#email
if "email" in df.columns:
    df["email"] = df["email"].str.lower()

    def valid_email(x):
        if pd.isna(x):
            return np.nan
        return x if re.match(r"^[\w\.-]+@[\w\.-]+\.\w+$", x) else np.nan

    df["email"] = df["email"].apply(valid_email)

# phone
if "phone" in df.columns:
    def clean_phone(x):
        if pd.isna(x):
            return np.nan
        x = re.sub(r"\D", "", str(x))  # only digits
        if x.startswith("998"):
            return "+" + x
        elif x.startswith("0"):
            return "+998" + x[1:]
        return np.nan

    df["phone"] = df["phone"].apply(clean_phone)
```

Step 5: JSON parsing
JSON columnlarini ochish (profile_json).
JSON ichidagi ma’lumotlarni alohida columnlarga ajratish: hobbies, skills, family, devices.
JSON ichidagi array yoki nested objectlarni tekshirish va flatten qilish.

```python
if "profile_json" in df.columns:

    def parse_json(x):
        try:
            return json.loads(x)
        except:
            return {}

    parsed = df["profile_json"].apply(parse_json)

    df["hobbies"] = parsed.apply(lambda x: x.get("hobbies"))
    df["skills"] = parsed.apply(lambda x: x.get("skills"))
    df["family"] = parsed.apply(lambda x: x.get("family"))
    df["devices"] = parsed.apply(lambda x: x.get("devices"))

```

Step 6: Address parsing
Address columnni alohida qism va key-larga ajratish: shahar, tuman, pochta kodi.
Raw address columnni saqlab, yangi columnlar hosil qilish (addr_city, addr_district, addr_postal).

```python
if "address" in df.columns:
    def split_address(x):
        if pd.isna(x):
            return pd.Series([np.nan, np.nan, np.nan])
        parts = x.split(",")
        return pd.Series([
            parts[0] if len(parts) > 0 else np.nan,
            parts[1] if len(parts) > 1 else np.nan,
            parts[2] if len(parts) > 2 else np.nan
        ])

    df[["addr_city", "addr_district", "addr_postal"]] = df["address"].apply(split_address)

```

Step 7: Duplicate va missing data tekshirish
Duplicate rowsni aniqlash va o‘chirish.
Muhim columnlarda missing values mavjudligini aniqlash va qaror qabul qilish (fillna yoki dropna).

```python
df = df.dropna(subset=["email"]) 
```

```python
df = df.drop_duplicates(subset=['student_id', 'name', 'email'])
```

Step 8: Data normalization
Gender columnini standard formatga keltirish: Male / Female / Unknown.
Course columnlarini yagona nomga keltirish: Data Science / Python / Other.
Status columnlarini standart formatga keltirish (lower-case yoki uniform).

```python
df = df.dropna(subset=["email"])  # xohlasang o‘zgartir


# STEP 8: Normalization
if "gender" in df.columns:
    df["gender"] = df["gender"].str.lower()
    df["gender"] = df["gender"].replace({
        "male": "Male",
        "m": "Male",
        "female": "Female",
        "f": "Female"
    })
    df["gender"] = df["gender"].fillna("Unknown")


if "course" in df.columns:
    df["course"] = df["course"].str.lower()

    def normalize_course(x):
        if "data" in x:
            return "Data Science"
        elif "python" in x:
            return "Python"
        else:
            return "Other"

    df["course"] = df["course"].apply(normalize_course)


if "status" in df.columns:
    df["status"] = df["status"].str.lower()
```

Barcha columnlarning to‘g‘ri type da ekanligini tekshirish: string, int, float, datetime.
